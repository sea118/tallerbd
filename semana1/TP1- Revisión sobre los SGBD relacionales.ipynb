## 1. Transformación del Modelo Entidad-Asociación hacia Modelo Relacional

Proponer un modelo relacional que corresponde al modelo Entidad-Asociación.

Medios_Prensa: {PK.NUmero_followers,Dueño,Nombre}
Noticias: {PK.ID,texto,titulo,FK.fecha}
    

## 2. Crear la estructura de la base de datos desde un programa Python

### 2.1 Conectarse a MySQL desde Python

```python
import mysql.connector
db1=mysql.connector.connect(host="localhost", user="root",passwd="123")
cursor=db1.cursor()
try:
    cursor.execute("CREATE DATABASE tp1")
except Exception as excepcion:
    print(excepcion)
cursor.execute("use tp1;")


```

### 2.2 Crear las tablas

```python
creatabla1="""Create table Medios_Prensa(nombre varchar(30) not null primary key, numero_followers int, dueño text);"""
creatabla2="""Create table Noticias(ID int not null auto_increment primary key, Titulo text, texto text ,fecha datetime,nombre varchar(30), FOREIGN KEY(nombre) references Medios_Prensa(nombre));"""
try:
    cursor.execute(creatabla1)
except Exception as excepcion:
    print(excepcion)
try:
    cursor.execute(creatabla2)
except Exception as excepcion:
    print(excepcion)
```

## 3.2 Leer los archivos textos y insertar los datos

```python
import csv, operator

with open('medios.csv') as csvarchivo:
    entrada = csv.reader(csvarchivo)
    saltalinea=False
    for reg in entrada:
        if(saltalinea):
            try:
                nombre=reg[0]
                cursor.execute("insert into Medios_Prensa(nombre,dueño,numero_followers) values('"+nombre.replace(" ","").replace("/n","")[0:29]+"','"+reg[1].replace(" ","")+"','"+reg[2].replace("/n","")+"');")
            except:
                print("Se agrego otra llave primaria en la tabla medios_prensa")
                break
        else:
            saltalinea=True
cursor.execute("SET FOREIGN_KEY_CHECKS = 0;")
cursor.execute("ALTER TABLE Noticias DEFAULT CHARSET=utf8 COLLATE=utf8_spanish_ci;")
archivo = open("noticias.txt","r")
a=0
fecha=""
nombre=""
texto=""
titulo=""
for linea in archivo.readlines():
    if(a==0):
        fecha=linea.replace("\n","")
        a+=1
    elif(a==1):
        nombre=linea.replace(" ","").replace("\n","")[0:29]
        a+=1
    elif(a==2):
        titulo=linea
        a+=1
    elif(a==3 and linea[0]!="-" and linea[1]!="-" and linea[2]!="-"):
        texto=texto+linea
    else:
        Query="insert into Noticias(fecha,nombre,Titulo,texto) values('"+fecha+"','"+nombre+"','"+titulo+"','"+texto+"');"
        cursor.execute(Query)
        a=0
        texto=""
archivo.close() 
```

# 4. Consultar la base de datos

Listar los medios de prensa en Chile ordenado según su número de followers en Twitter.

```python
cursor.execute("select nombre,numero_followers from Medios_Prensa order by numero_followers;")
filas = cursor.fetchall()
for fila in filas:
    print(fila)
    
```

```python
cursor.execute("select count(*) from (select dueño from Medios_Prensa group by dueño) as dueños;")
filas = cursor.fetchall()
for fila in filas:
    print(fila)
```

Contar cuantes medios de prensa posee cada dueño.

```python
cursor.execute("select dueño ,count(*) from Medios_Prensa group by dueño;")
filas = cursor.fetchall()
for fila in filas:
    print(fila)
```

Mostrar los últimos 10 títulos de noticias del medio "La Tercera"

```python
cursor.execute("select Titulo from Noticias where nombre like '%tercera%' order by fecha desc limit 10;")
filas = cursor.fetchall()
for fila in filas:
    print(fila)
```

Mostrar los nombre de medios y los títulos en los cuales aparacen la palabra "Valdivia".

```python
cursor.execute("select Titulo,nombre from Noticias where nombre like '%valdivia%' or Titulo like '%valdivia%';")
filas = cursor.fetchall()
for fila in filas:
    print(fila)
```

Por cada medio, mostrar el número de títulos en los cuales aparecen la palabra "Puerto Montt".

```python
cursor.execute("select nombre, count(*) from Noticias where Titulo like '%Puerto Montt%' group by nombre;")
filas = cursor.fetchall()
for fila in filas:
    print(fila)
```
