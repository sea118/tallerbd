```python
consulta3=twitter.search(q='valdivia', count=10, geocode='-39.8213966,-73.2467844,1mi')
db = connection['final']
final = db.consulta3
for result in consulta3["statuses"]:
    print("(%s) @%s %s %s" % (result["created_at"], result["user"]["screen_name"], result["text"],result["user"]["location"]))
    db.final.insert_one({'create_at':result["created_at"],'username':result["user"]["screen_name"],'text':result["text"],'location':result["user"]["location"]})
    print("se inserto nuevo elemento" +'\n')
```

```python
cursor=db.final.find()
for element in cursor:
    print(element['location'])
```

```python
cursor=db.final.find({"text":"/UACh/"}) 
for element in cursor:
    print(element)
```

```python
cursor=db.final.aggregate([{"$group":{"_id":"$username","count":{"$sum":1}}}])
for elemento in cursor:
    print(elemento)
```
