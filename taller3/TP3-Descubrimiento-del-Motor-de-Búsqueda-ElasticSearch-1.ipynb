```python
#insertar otro documento de prueba

from datetime import datetime
from elasticsearch import Elasticsearch
es = Elasticsearch()
doc ={
    "user": "jorge obando",
    "post_date" : "JAN 2018",
    "resumen" : "este es un resumen de prueba",
}
res = es.index(index="articulo-uach", doc_type='_doc', id=6, body=doc)
```

```python
#consulta palabra clave
import pprint
final=es.search(index="articulo-uach", body={"query": {"match": {'resumen':'Manganese'}}})
pprint.pprint(final)
```
